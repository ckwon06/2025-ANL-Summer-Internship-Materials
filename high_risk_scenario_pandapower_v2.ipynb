```python
import pandas as pd

def compare_scenarios(base_net, net1, net2, label_base, label_1, label_2):
    # Align lengths if needed
    def align(df1, df2, df3):
        max_len = max(len(df1), len(df2), len(df3))
        return (
            df1.reindex(range(max_len)),
            df2.reindex(range(max_len)),
            df3.reindex(range(max_len))
        )

    # Line data comparison
    line_base, line_1, line_2 = align(base_net.res_line, net1.res_line, net2.res_line)
    line_df = pd.concat([
        line_base.add_prefix(f"{label_base}_"),
        line_1.add_prefix(f"{label_1}_"),
        line_2.add_prefix(f"{label_2}_")
    ], axis=1)

    # Bus data comparison
    bus_base, bus_1, bus_2 = align(base_net.res_bus, net1.res_bus, net2.res_bus)
    bus_df = pd.concat([
        bus_base.add_prefix(f"{label_base}_"),
        bus_1.add_prefix(f"{label_1}_"),
        bus_2.add_prefix(f"{label_2}_")
    ], axis=1)

    return line_df, bus_df

# Compare each group
line_9, bus_9 = compare_scenarios(net0, net1a, net2a, "0", "1a", "2a")
line_30, bus_30 = compare_scenarios(net00, net1b, net2b, "00", "1b", "2b")
line_118, bus_118 = compare_scenarios(net000, net1c, net2c, "000", "1c", "2c")

# Display tables
print("=== 9-BUS SYSTEM COMPARISON ===")
display(line_9)
display(bus_9)

print("=== 30-BUS SYSTEM COMPARISON ===")
display(line_30)
display(bus_30)

print("=== 118-BUS SYSTEM COMPARISON ===")
display(line_118)
display(bus_118)
```

Here are three tables comparing the line and bus data of standard 9, 30, and 118 bus power systems to similar 9, 30, and 118 bus modified into scenarios 1 and 2 from the previous experiment. Scenario 1 overloads bus 5 by increasing active power while Scenario 2 trips the generator by setting its power to 0.
